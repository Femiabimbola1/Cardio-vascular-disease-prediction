```python
#check for outliers
cols = heart.columns

Q1 = heart[cols].quantile(0.25)
Q3 = heart[cols].quantile(0.75)
IQR = Q3 - Q1

threshold = 1.5

index = np.where((heart[cols] < (Q1 - threshold * IQR)) | (heart[cols] > (Q3 + threshold * IQR)))[0]

heart = heart.drop(heart.index[index])
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
from skopt import BayesSearchCV
```

```python
rfc_model = RandomForestClassifier()

param_grid = {
    'n_estimators':[100,200,300],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'max_features':['auto','sqrt','log2'],
    'bootstrap':[True, False]
}

bayes_search = BayesSearchCV(rfc_model,param_grid, cv=5,error_score ='raise' )
bayes_search.fit(X_train, y_train)

print('Best Hyperparameters:',bayes_search.best_params_)
print('Best Score(Accuracy):',bayes_search.best_score_)
```

```python
Best_hyperparameter_rfc = {'bootstrap': False, 'max_depth': None, 'max_features':'log2', 'min_samples_leaf': 2, 'min_samples_split': 2, 'n_estimators': 300}
```

```python
rfc_model = RandomForestClassifier(**Best_hyperparameter_rfc)

rfc_model.fit(X_train, y_train)

rfc_model
```

```python
rfc_prediction = rfc_model.predict(X_test)

rfc_prediction[0:5]
```

**XGB Classifier**

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
xgb_model = XGBClassifier()

param_grid = {
    'n_estimators':[100,200,300],
    'max_depth':[3,5,7],
    'learning_rate':[0.01,0.1,0.2],
    'subsample':[0.8,0.9,1.0],
    'colsample_bytree':[0.8,0.9,1.0],
    'gamma':[0,0.1,0.2],
    'min_child_weight':[1,2,3]
}

grid_search = GridSearchCV(xgb_model, param_grid, cv=5, error_score='raise')
grid_search.fit(X_train, y_train)

print('Best Hyperparameter:',grid_search.best_params_)
print('Best Score(Accuracy);', grid_search.best_score_)
```

```python
Best_Hyperparameter_xgb = {'colsample_bytree': 0.9, 'gamma': 0.2, 'learning_rate': 0.2, 'max_depth': 7, 'min_child_weight': 2, 'n_estimators': 100, 'subsample': 0.8}
```

```python
xgb_model = XGBClassifier(**Best_Hyperparameter_xgb)

xgb_model.fit(X_train, y_train)

xgb_model
```

```python
xgb_prediction = xgb_model.predict(X_test)

xgb_prediction[0:5]
```

**Model Evaluation**

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 7))

sns.heatmap(confusion_matrix(y_test, rfc_prediction), annot=True, ax=axes[0])
axes[0].set_title('Random Forest',fontsize=18)

sns.heatmap(confusion_matrix(y_test, xgb_prediction), annot=True, ax=axes[1])
axes[1].set_title('XGBoost Classifier',fontsize=18)

sns.heatmap(confusion_matrix(y_test, tree_prediction), annot=True, ax=axes[2])
axes[2].set_title('Decision Tree', fontsize=18)

for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), ha='right', fontsize=18)

plt.tight_layout()
plt.show()
```

```python
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    ax = plt.subplot()
    sns.heatmap(cm, annot=True, ax=ax)
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('Actual labels')
    ax.set_title(f'Confusion Matrix - {model_name}')
    ax.xaxis.set_ticklabels(['Heart Disease Absent', 'Heart Disease Present'])
    ax.yaxis.set_ticklabels(['Heart Disease Absent', 'Heart Disease Present'])
    plt.show()
```

```python
y = y_test
y_predict= xgb_prediction
model_name = 'XGBClassifier'
plot_confusion_matrix(y,y_predict,model_name)
```

```python
precision_xgb = precision_score(y_test, xgb_prediction)
recall_xgb = recall_score(y_test, xgb_prediction)
f1_xgb = f1_score(y_test, xgb_prediction)
acc_xgb = xgb_model.score(X_test, y_test)

precision_rf = precision_score(y_test, rfc_prediction)
recall_rf = recall_score(y_test, rfc_prediction)
f1_rf = f1_score(y_test, rfc_prediction)
acc_rf = rfc_model.score(X_test, y_test)

precision_dt = precision_score(y_test, tree_prediction)
recall_dt = recall_score(y_test, tree_prediction)
f1_dt = f1_score(y_test, tree_prediction)
acc_dt = tree_model.score(X_test, y_test)
```

```python

list_precision = [precision_rf,precision_xgb,precision_dt]
list_recall = [recall_rf,recall_xgb,recall_dt]
list_f1 = [f1_rf,f1_xgb,f1_dt]
list_acc = [acc_rf,acc_xgb,acc_dt]

Report = pd.DataFrame(list_precision, index=['RandomForest Classifier','XGB Classifier','Decision Tree'])
Report.columns =['Precision']
Report.insert(loc=1,column='Recall',value=list_recall)
Report.insert(loc=2, column='F1 Score', value=list_f1)
Report.insert(loc=3, column='Accuracy', value=list_acc)
Report.columns.name = 'Algorithm'
print(Report)
```
